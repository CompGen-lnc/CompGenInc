```python
import pandas as pd
```

```python
# Leer archivos << ss_loops_dict se lee para saber a cuales archivos generar el .fa de seq y 2d.
# data_miRNA contiene Nombre miRNA,Secuencia miRNA Especifico,Nombre Precursor,Secuencia Precursor,Secuencia 2D Precursor
# Para distintos miRNA

ss_loops_dict = pd.read_csv("./../data/fasta/fasta_seq/miRNA.fa", header=None)
data_miRNA = pd.read_csv("./../data/data_miRNA.txt")

# Extraer las columnas que se van a necesitar
nombre_miRNA = data_miRNA['Nombre miRNA'].tolist()
secuencia_2d = data_miRNA['Secuencia 2D Precursor'].tolist()
secuencia_miRNA = data_miRNA['Secuencia miRNA Especifico'].tolist()
secuencia_precursor = data_miRNA['Secuencia Precursor'].tolist()
```

```python
# Extraer nombres de miRNA

test_mir = []
for i in range(0, len(ss_loops_dict), 2):
    mirna = ss_loops_dict.iloc[i, 0].split('>')[1]
    test_mir.append(mirna)
```

```python
# Definir listas donde almacenar la informacion que luego se va a escribir
miRNA_TRUE=[]
miRNA_FALSE=[]
miRNA_2d=[]
miRNA_seq=[]

# Agregar breve descripcion al archivo que se genera
miRNA_TRUE.append("En este .txt se pueden encontrar los miRNA para los cuales se SI generaron los archivos")
miRNA_FALSE.append("En este .txt se pueden encontrar los miRNA para los cuales se NO generaron los archivos")
```

```python
for i in range(len(test_mir)):
    if test_mir[i] in nombre_miRNA:
        indice = nombre_miRNA.index(test_mir[i])
        secuencia_precursor_str=str(secuencia_precursor[indice])
        secuencia_miRNA_str=str(secuencia_miRNA[indice])
        secuencia_2d_str=str(secuencia_2d[indice])
        miRNA_TRUE.append(test_mir[i])
        miRNA_2d.append(f'>{test_mir[i]}')
        miRNA_2d.append(secuencia_2d_str)
        miRNA_seq.append(f'>{test_mir[i]}')
        miRNA_seq.append(secuencia_precursor[indice])
    else:
        miRNA_FALSE.append(test_mir[i])
```

```python
# Guardar la informacion que se obtuvo

# miRNA para los cuales se genero seq y 2d

output_file_path = "./../data/miRNA_TRUE.txt"
with open(output_file_path, "w") as f:
    for i in range(len(miRNA_TRUE)):
        f.write(f'{miRNA_TRUE[i]}\n')

# miRNA para los cuales no se genero seq y 2d

output_file_path = "./../data/miRNA_FALSE.txt"
with open(output_file_path, "w") as f:
    for i in range(len(miRNA_FALSE)):
        f.write(f'{miRNA_FALSE[i]}\n')

# .fa para miRNA 2d
output_file_path = "./../data/miRNA_2d.fa"
with open(output_file_path, "w") as f:
    for i in range(len(miRNA_2d)):
        f.write(f'{miRNA_2d[i]}\n')

# .fa para miRNA seq

output_file_path = "./../data/miRNA_seq.fa"
with open(output_file_path, "w") as f:
    for i in range(len(miRNA_seq)):
        f.write(f'{miRNA_seq[i]}\n')
```
